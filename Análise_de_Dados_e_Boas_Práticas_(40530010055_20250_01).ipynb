## üîç Detec√ß√£o de Outliers em S√©ries Temporais

### ‚úîÔ∏è O que s√£o Outliers?

Outliers s√£o valores **fora do padr√£o esperado**, que se distanciam significativamente da m√©dia ou do comportamento hist√≥rico da s√©rie.

---

### üî∏ Por que eles ocorrem em dados financeiros?

- üì∞ Eventos inesperados (guerras, pandemias, crises pol√≠ticas).
- üí∏ Alta volatilidade.
- üè¶ Decis√µes de bancos centrais (juros, pol√≠tica monet√°ria).
- ‚öôÔ∏è Problemas t√©cnicos ou erros nas fontes de dados.

---

## üìä M√©todos comuns para detectar outliers em s√©ries temporais:

### 1Ô∏è‚É£ **Regra do Desvio Padr√£o**

Se um valor est√° acima de **¬±3 desvios-padr√£o** da m√©dia, √© considerado um outlier.

---

### 2Ô∏è‚É£ **Boxplot (M√©todo dos Quartis)**

Detecta outliers fora dos limites:

- üîπ **Limite Superior:** Q3 + 1.5 * IQR  
- üîπ **Limite Inferior:** Q1 - 1.5 * IQR

---

### 3Ô∏è‚É£ **Detec√ß√£o Visual**

- üìà Gr√°ficos ajudam a ver pontos fora do padr√£o.  
- Usar linhas de pre√ßo, rolling mean, ou boxplots.

---

## ‚ö†Ô∏è Aten√ß√£o Importante:

- **Outliers nem sempre s√£o erros.**  
Em s√©ries financeiras, eles muitas vezes representam **eventos reais de mercado.**

- Remover outliers **depende do contexto:**  
Se s√£o erros, remove-se.  
Se representam movimentos extremos (como quedas ou disparadas de pre√ßo), **devem ser mantidos**, pois s√£o relevantes para modelos financeiros.

---

## üö¶ Devo tratar outliers agora?

- ‚ùå **N√£o.**  
Na fase atual (an√°lise de dados, constru√ß√£o dos √≠ndices locais e globais), **os outliers fazem parte da din√¢mica dos mercados.**

- ‚úÖ A avalia√ß√£o sobre tratar ou n√£o ser√° feita na fase de modelagem da IA, especialmente se formos treinar modelos sens√≠veis a esses extremos.

---

## üéØ Conclus√£o:

- ‚úîÔ∏è Detectar outliers √© relevante para entender eventos extremos na s√©rie.
- ‚úîÔ∏è Por enquanto, eles **n√£o devem ser removidos**, mas sim observados como parte natural dos movimentos do mercado.

## ‚öñÔ∏è Balanceamento de Dados em S√©ries Temporais

### ‚úîÔ∏è O que √© balanceamento de dados?

Em problemas tradicionais de **classifica√ß√£o**, o balanceamento se refere √† propor√ß√£o equilibrada entre as classes (ex.: 50% compra, 50% venda). Se uma classe aparece muito mais que outra, dizemos que o dataset est√° **desbalanceado**, o que pode prejudicar o desempenho dos modelos de machine learning.

---

## üö´ **Mas isso se aplica a s√©ries temporais?**

**N√£o da mesma forma.**  
S√©ries temporais possuem uma estrutura dependente do tempo, onde a ordem dos dados √© fundamental. Portanto:

- ‚ùå **N√£o faz sentido embaralhar os dados.**  
- ‚ùå N√£o se pode aplicar t√©cnicas tradicionais de oversampling (SMOTE, ROS) ou undersampling.

---

## üîç E como avaliar o "balanceamento"?

O que se pode observar em s√©ries temporais √©:

- Frequ√™ncia dos eventos (ex.: quantas vezes aparece um sinal de compra, venda ou neutro em determinado per√≠odo).
- Se o mercado passou por longos per√≠odos de alta, queda ou lateraliza√ß√£o, isso **reflete o pr√≥prio comportamento do mercado naquele per√≠odo**, e **n√£o √© um problema de dataset desbalanceado.**

---

## ‚úÖ Quando o balanceamento vira uma preocupa√ß√£o?

Na fase de modelagem preditiva, se formos transformar os sinais em classes (ex.: compra, venda, neutro), podemos verificar:

- üî∏ Se o mercado tende mais a um tipo de cen√°rio (ex.: mercado direcional, mais compras ou mais vendas).
- üî∏ Se h√° poucos exemplos de uma classe (ex.: raramente aparece venda), o modelo pode ficar enviesado.

Neste caso, algumas abordagens s√£o v√°lidas:

- üìä **Ajuste de pesos nas classes.**
- üîÑ **Cria√ß√£o de janelas temporais diferentes (ex.: observar cen√°rios semanais ou mensais).**
- üö¶ Ou simplesmente **aceitar que o mercado, por natureza, n√£o √© balanceado**, e treinar o modelo para aprender esse comportamento.

---

## üìå Importante para este MVP:

- ‚úÖ N√£o estamos aplicando balanceamento agora, pois esta √© a etapa de **an√°lise explorat√≥ria e constru√ß√£o dos indicadores.**
- ‚úÖ O comportamento dos dados reflete exatamente como o mercado se comportou no per√≠odo analisado.
- üöÄ O balanceamento, se necess√°rio, ser√° avaliado na etapa de **treinamento dos modelos de IA.**

---

## üéØ Conclus√£o:

- S√©ries temporais **n√£o precisam ser balanceadas no sentido tradicional.**
- O foco est√° em entender os padr√µes sequenciais e o comportamento din√¢mico dos mercados.

## üîç Diferen√ßa entre S√©ries Temporais e Dados Tabulares Tradicionais

### üóÇÔ∏è Dados Tabulares Tradicionais

- Cada linha representa uma **amostra independente**.
- A ordem das linhas **n√£o importa**.
- As colunas s√£o vari√°veis ou atributos do problema.
- Exemplos: dados de clientes, transa√ß√µes, exames m√©dicos, imagens (quando vetorizadas).

‚úîÔ∏è Os modelos assumem que as amostras s√£o **independentes e identicamente distribu√≠das (i.i.d)**.

---

### ‚è≥ S√©ries Temporais

- Cada linha representa uma **observa√ß√£o em um instante no tempo**.
- A ordem dos dados √© **essencial e obrigat√≥ria**.
- As colunas representam as vari√°veis que evoluem no tempo (pre√ßo, volume, indicadores, etc.).
- Exemplos: pre√ßos de ativos financeiros, sensores IoT, clima, vendas di√°rias.

‚úîÔ∏è As amostras **n√£o s√£o independentes**. Existe depend√™ncia temporal entre os dados anteriores e os futuros.

---

### üö¶ Principais diferen√ßas pr√°ticas:

| Aspecto                | Dados Tabulares         | S√©ries Temporais             |
| ---------------------- | ----------------------- | ---------------------------- |
| Ordem dos dados        | N√£o importa             | Importante e obrigat√≥ria     |
| Depend√™ncia temporal   | N√£o existe              | Existe (dados sequenciais)   |
| T√©cnicas de modelagem  | Random Forest, XGBoost, Redes Neurais Feedforward | ARIMA, LSTM, Prophet, Modelos sequenciais |
| Cross-validation       | Aleat√≥ria (k-fold)      | Baseada em tempo (forward chaining) |
| Balanceamento          | Avaliado diretamente nas classes | Depende do comportamento do mercado no tempo |
| Predi√ß√£o               | Classe ou valor isolado | Sequ√™ncia ou pr√≥ximo valor no tempo |

---

### üéØ Conclus√£o:

- Os dados s√£o s√©ries temporais de ativos financeiros.
- Toda an√°lise, modelagem e interpreta√ß√£o **leva em considera√ß√£o o fator tempo.**
- Balanceamento, valida√ß√£o cruzada e treinamento s√£o feitos **de forma sequencial, preservando a ordem cronol√≥gica.**

- üî• Modelar s√©ries temporais √© muito mais que simplesmente tratar dados tabulares.  
- ‚öôÔ∏è Exige o uso de t√©cnicas espec√≠ficas que consideram o **comportamento din√¢mico dos mercados ao longo do tempo.**

## üîß Normaliza√ß√£o dos Dados

### üéØ Objetivo:
Garantir que todos os ativos estejam na **mesma escala**, facilitando compara√ß√µes visuais, an√°lises estat√≠sticas e aplica√ß√£o de modelos.

### üîç Por que normalizar?
- Os ativos possuem **escalas muito diferentes**. Por exemplo:
  - WINFUT trabalha na faixa de centenas de milhares de pontos.
  - DOLFUT opera na casa dos 5 a 6 reais.
  - A√ß√µes como VALE3, PETR4, ITUB4 possuem pre√ßos unit√°rios diversos.
- Sem normaliza√ß√£o, ativos de maior valor absoluto podem **dominar an√°lises visuais ou c√°lculos estat√≠sticos**, mascarando os movimentos relativos dos demais.

### ‚öôÔ∏è M√©todo aplicado:
- **Normaliza√ß√£o Min-Max**, onde cada valor √© transformado segundo a f√≥rmula:

\[
\text{Valor Normalizado} = \frac{\text{Valor Atual} - \text{M√≠nimo}}{\text{M√°ximo} - \text{M√≠nimo}}
\]

- Resultado:
  - Todos os ativos s√£o escalados para uma faixa de **0 a 1**.
  - Preserva a proporcionalidade dos movimentos.

### üöÄ Aplica√ß√µes desta etapa:
- Gera√ß√£o de gr√°ficos comparativos de ativos com escalas diferentes.
- Constru√ß√£o de indicadores compostos.
- Base para algoritmos que s√£o sens√≠veis √† escala (ex.: KNN, redes neurais, regress√µes, etc.).

---


```python
# ===============================================
# üîß Normaliza√ß√£o dos dados entre 0 e 1
# ===============================================

# A normaliza√ß√£o √© uma etapa importante para deixar todos os ativos na mesma escala,
# facilitando compara√ß√µes, an√°lises estat√≠sticas e modelos de machine learning.

# F√≥rmula aplicada:
# Normalizado = (valor - valor_m√≠nimo) / (valor_m√°ximo - valor_m√≠nimo)

df_normalizado = (df_fechamento - df_fechamento.min()) / (df_fechamento.max() - df_fechamento.min())

# ‚úÖ Exibir os primeiros dados normalizados
print("\nüìä Dados Normalizados (amostra):")
print(df_normalizado.head())
```

## üîç Por que utilizar dados normalizados?

Nesta etapa de estat√≠sticas descritivas, foi aplicada a normaliza√ß√£o dos dados entre 0 e 1, utilizando a f√≥rmula:

$$
X_{\text{normalizado}} = \frac{X - X_{\min}}{X_{\max} - X_{\min}}
$$

‚úîÔ∏è A normaliza√ß√£o permite:

- Comparar ativos com escalas muito diferentes (ex.: WINFUT na casa dos 120 mil pontos e DOLFUT na casa de 5).
- Analisar a dispers√£o dos dados de forma proporcional ao comportamento hist√≥rico de cada ativo.
- Observar varia√ß√µes relativas em vez de valores absolutos.

‚ö†Ô∏è Importante destacar que esta normaliza√ß√£o √© usada exclusivamente para **an√°lise explorat√≥ria e interpreta√ß√£o estat√≠stica dos dados**, e n√£o altera os dados originais para outros c√°lculos que dependem de pre√ßos reais.

‚û°Ô∏è Esta abordagem ser√° tamb√©m extremamente √∫til nas pr√≥ximas etapas de desenvolvimento dos modelos de machine learning, onde muitos algoritmos s√£o sens√≠veis √† escala dos dados.

## üîó Import√¢ncia da Matriz de Correla√ß√£o no Contexto do MVP

### üìå O que √© uma Matriz de Correla√ß√£o?

A matriz de correla√ß√£o √© uma ferramenta estat√≠stica que mede a rela√ß√£o linear entre dois ou mais ativos financeiros. Ela quantifica o grau de associa√ß√£o entre as vari√°veis, com valores que variam de **-1** a **+1**:

- ‚úÖ **+1:** Correla√ß√£o perfeitamente positiva (os ativos se movem na mesma dire√ß√£o).
- ‚úÖ **0:** Sem correla√ß√£o linear.
- ‚úÖ **-1:** Correla√ß√£o perfeitamente negativa (os ativos se movem em dire√ß√µes opostas).

---

### üéØ Por que √© Importante no Nosso MVP?

1. **üß† Entendimento do Mercado:**
   - Permite observar como ativos locais e globais est√£o inter-relacionados.
   - Avalia se o mercado est√° mais sens√≠vel ao cen√°rio interno (Brasil) ou externo (EUA, d√≥lar, juros).

2. **üìä Constru√ß√£o de Indicadores Sint√©ticos:**
   - A matriz valida se os ativos selecionados t√™m rela√ß√£o forte entre si, garantindo que o **√çndice Sint√©tico Local e Global** sejam representativos.

3. **üö¶ Gera√ß√£o de Sinais Operacionais:**
   - Correla√ß√µes fortes ajudam a definir contextos favor√°veis ou desfavor√°veis para opera√ß√µes.
   - Ex.: Alta correla√ß√£o entre WINFUT e ativos locais ‚Üí **mercado direcional e forte.**

4. **‚öôÔ∏è Robustez no Modelo Quantitativo:**
   - Reduz ru√≠do. Selecionar ativos com alta correla√ß√£o no grupo local ou global ajuda a evitar sinais inconsistentes.
   - Serve de input ou feature relevante para IA e machine learning.

5. **üîç An√°lise de Sensibilidade:**
   - A matriz explica quais ativos mais impactam o Mini √çndice (WINFUT) ou o Mini D√≥lar (DOLFUT).
   - Permite construir modelos ponderados mais pr√≥ximos da realidade do mercado.

---

### üö® Riscos de N√£o Considerar a Correla√ß√£o:

- ‚ùå Gerar modelos fr√°geis e inconsistentes.
- ‚ùå Ignorar influ√™ncia externa (ex.: d√≥lar, juros EUA) sobre ativos locais.
- ‚ùå Operar contra o contexto macroecon√¥mico, aumentando riscos.

---

### üí° Conclus√£o:

A matriz de correla√ß√£o n√£o √© apenas uma an√°lise estat√≠stica, mas uma **ferramenta estrat√©gica fundamental** para compreender o comportamento dos ativos e aumentar a assertividade dos modelos e opera√ß√µes.  

Ela conecta o **contexto macroecon√¥mico global e local** diretamente √†s decis√µes operacionais no mercado futuro (**WINFUT** e **DOLFUT**).


```python
# ==============================================
# üî• Matriz de Correla√ß√£o dos Ativos Locais (IBOV)
# ==============================================

# ‚úÖ Seleciona os ativos locais (valores do dicion√°rio)
ativos_selecionados = list(ativos_locais.values())

# ‚úÖ Calcula os retornos percentuais
retornos = df_fechamento.pct_change().dropna()

# ‚úÖ Calcula a matriz de correla√ß√£o
corr_local = retornos[ativos_selecionados].corr()

# ‚úÖ Cria o heatmap da matriz de correla√ß√£o
plt.figure(figsize=(10, 8))
sns.heatmap(
    corr_local,
    annot=True,
    cmap='coolwarm',
    fmt=".2f",
    linewidths=0.5,
    linecolor='white',
    cbar=True,
    square=True,
    annot_kws={"size": 10}
)

plt.title('Matriz de Correla√ß√£o ‚Äî Ativos Locais (IBOV)', fontsize=16)
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# ==========================
# üîó Dados dos grupos
# ==========================

grupos = {
    'Setor Financeiro (Bancos)': ['ITUB4', 'BBDC4', 'BBAS3'],
    'IBOV (B3SA3)': ['B3SA3'],
    'Commodities (VALE/PETR)': ['VALE3', 'PETR4', 'PETR3'],
    'Mercado Global (SP500/Dow)': ['SP500', 'DOWJONES'],
    'DXY (√çndice D√≥lar)': ['DXY'],
    'Treasury10Y (Juros EUA)': ['TREASURY10Y']
}

# ==========================
# üîç C√°lculo da Sensibilidade
# ==========================

janela = 21  # janela de 21 per√≠odos

sensibilidade_win = []
sensibilidade_dol = []

for nome_grupo, lista_ativos in grupos.items():
    # Calcula a correla√ß√£o m√≥vel m√©dia dos ativos do grupo com WINFUT
    corr_win_media = np.mean([
        df_fechamento['WINFUT'].pct_change().rolling(janela).corr(df_fechamento[ativo].pct_change()).mean()
        for ativo in lista_ativos
    ])

    # Calcula a correla√ß√£o m√≥vel m√©dia dos ativos do grupo com DOLFUT
    corr_dol_media = np.mean([
        df_fechamento['DOLFUT'].pct_change().rolling(janela).corr(df_fechamento[ativo].pct_change()).mean()
        for ativo in lista_ativos
    ])

    sensibilidade_win.append(round(corr_win_media, 2))
    sensibilidade_dol.append(round(corr_dol_media, 2))

# ==========================
# üî• Plot do Painel de Sensibilidade
# ==========================

x = np.arange(len(grupos))  # posi√ß√µes das barras

plt.figure(figsize=(12,6))
bar_width = 0.35

plt.bar(x - bar_width/2, sensibilidade_win, bar_width, label='WINFUT', color='royalblue')
plt.bar(x + bar_width/2, sensibilidade_dol, bar_width, label='DOLFUT', color='seagreen')

plt.axhline(0, color='black', linewidth=0.8)

plt.ylabel('Sensibilidade ‚Äî Correla√ß√£o M√©dia (21 dias)')
plt.title('Painel de Sensibilidade dos Ativos ‚Äî WINFUT vs DOLFUT')
plt.xticks(x, list(grupos.keys()), rotation=45, ha='right')

plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()

# ‚úÖ Verificar os valores num√©ricos
for grupo, win, dol in zip(grupos.keys(), sensibilidade_win, sensibilidade_dol):
    print(f'{grupo}: WINFUT={win}, DOLFUT={dol}')
```

1. **Defini√ß√£o dos Grupos Econ√¥micos:**

| Grupo                          | Ativos                                      |

### üî¢ F√≥rmula da Sensibilidade (Correla√ß√£o M√≥vel de 21 dias)

### üî¢ F√≥rmula da Sensibilidade (Correla√ß√£o M√≥vel de 21 dias)

$$
\text{Sensibilidade}_{\text{grupo}} = \frac{\sum_{i=1}^{n} \overline{\text{Corr}_{21}(\text{Ativo}_i, \text{Contrato})}}{n}
$$

**Onde:**

- $\text{Corr}_{21}$ = Correla√ß√£o m√≥vel de 21 dias.
- $\text{Ativo}_i$ = Ativo pertencente ao grupo.
- $\text{Contrato}$ = WINFUT ou DOLFUT.
- $n$ = Quantidade de ativos no grupo.
- $\overline{\text{Corr}}$ = M√©dia da correla√ß√£o no per√≠odo.

‚úîÔ∏è Esta f√≥rmula representa a **m√©dia da correla√ß√£o m√≥vel de 21 dias** entre cada ativo do grupo e o contrato analisado (WINFUT ou DOLFUT).


‚úîÔ∏è Esta f√≥rmula representa a **m√©dia da correla√ß√£o m√≥vel de 21 dias** entre cada ativo do grupo e o contrato analisado (WINFUT ou DOLFUT).

3. **Visualiza√ß√£o:**

### üîß Interpreta√ß√£o dos Resultados:

- **Correla√ß√£o positiva forte (pr√≥ximo de 1):** O contrato tende a se movimentar na mesma dire√ß√£o do grupo.

- **Correla√ß√£o negativa (pr√≥ximo de -1):** O contrato tende a se mover na dire√ß√£o oposta ao grupo.

- **Pr√≥ximo de zero:** Indica pouca ou nenhuma rela√ß√£o.

## üî• An√°lise de Contexto ‚Äî Ativos Locais e Globais

### üèóÔ∏è Metodologia:
- C√°lculo da **matriz de correla√ß√£o di√°ria** entre todos os ativos do portf√≥lio.
- Extra√ß√£o da m√©dia da correla√ß√£o dos ativos locais e globais com o WINFUT.
- Classifica√ß√£o do cen√°rio de mercado com base nos seguintes crit√©rios:

| Correla√ß√£o M√©dia | Interpreta√ß√£o                                | Estrat√©gia                              |
|------------------|----------------------------------------------|------------------------------------------|
| üî• ‚â• 0.80        | Mercado Direcional                           | Operar Tend√™ncia                        |
| üüß 0.50 a < 0.80 | Mercado Neutro                               | Operar com Cautela                      |

### üîç Interpreta√ß√£o dos Resultados:

- ‚úîÔ∏è Se os ativos **Locais** est√£o muito correlacionados com o WINFUT, isso indica for√ßa no mercado interno ‚Äî cen√°rio favor√°vel para opera√ß√µes no Mini √çndice.

- ‚ùó Se ambos est√£o desalinhados ou com baixa correla√ß√£o, o mercado se encontra em um contexto incerto ou altamente vol√°til.

---

### üéØ Aplica√ß√µes Pr√°ticas:

- üìä Este tipo de an√°lise permite **adaptar a estrat√©gia operacional** de acordo com o contexto:

| Cen√°rio Locais | Cen√°rio Globais | Estrat√©gia                                      |
|----------------|------------------|-------------------------------------------------|
| üèπ Direcional  | üèπ Direcional     | Forte tend√™ncia, operar a favor da tend√™ncia.   |
| üî∏ Neutro      | üî∏ Neutro         | Estrat√©gias de curto prazo, revers√µes r√°pidas.  |
| ‚ö†Ô∏è Desordenado | ‚ö†Ô∏è Desordenado    | Foco em prote√ß√£o, opera√ß√µes em volatilidade.    |
| üèπ Direcional  | ‚ö†Ô∏è Desordenado    | Foco no contexto local, evitar influ√™ncia global.|
| ‚ö†Ô∏è Desordenado | üèπ Direcional     | Atento ao cen√°rio externo, evitar √≠ndice, foco no d√≥lar.|

---

‚úîÔ∏è **Conclus√£o:**  
Este painel fornece uma vis√£o clara sobre o alinhamento dos mercados **Locais** e **Globais**, permitindo tomar decis√µes mais informadas e robustas.

---

```python
# ===========================================
# üî• An√°lise de Contexto ‚Äî Ativos Locais e Globais
# ===========================================

# ‚úÖ Calcula retornos
retornos = df_fechamento.pct_change().dropna()

# ‚úÖ Calcula correla√ß√£o di√°ria com WINFUT
corr_diaria = retornos.corr()

# üîµ Lista dos ativos locais e globais (ajuste conforme suas vari√°veis)
ativos_locais = ['VALE3', 'PETR4', 'ITUB4', 'PETR3', 'BBAS3', 'BBDC4', 'B3SA3', 'ABEV3', 'WEGE3', 'RENT3']
ativos_globais = ['SP500', 'NASDAQ', 'DOWJONES', 'DXY', 'TREASURY10Y', 'PETROLEO', 'BITCOIN', 'MEXICO']

# ‚úÖ Calcula m√©dia das correla√ß√µes dos ativos locais e globais com WINFUT
corr_locais = corr_diaria.loc[ativos_locais, 'WINFUT'].mean()
corr_globais = corr_diaria.loc[ativos_globais, 'WINFUT'].mean()

print(f"üß† M√©dia Correla√ß√£o Ativos Locais com WINFUT: {corr_locais:.2f}")
print(f"üåé M√©dia Correla√ß√£o Ativos Globais com WINFUT: {corr_globais:.2f}")

# ‚úÖ Definir fun√ß√£o para classificar o cen√°rio
def classificar_cenario(correlacao):
    if correlacao >= 0.80:
        return 'üèπ Mercado Direcional ‚Äî Operar Tend√™ncia'
    elif 0.50 <= correlacao < 0.80:
        return 'üî∏ Mercado Neutro ‚Äî Operar com Cautela'
    else:
        return '‚ö†Ô∏è Mercado Desordenado ‚Äî Priorizar Prote√ß√£o / Volatilidade'

# ‚úÖ Aplicar a fun√ß√£o
cenario_locais = classificar_cenario(corr_locais)
cenario_globais = classificar_cenario(corr_globais)

# ‚úÖ Exibir o cen√°rio atual
print('\nüìä üîó Contexto dos Ativos Locais:', cenario_locais)
print('üìä üîó Contexto dos Ativos Globais:', cenario_globais)
```

### üîó F√≥rmula Aplicada:

O **√çndice Local** √© calculado utilizando a m√©dia ponderada dos retornos dos ativos, de acordo com seus pesos no √≠ndice IBOV.

---

#### üîπ C√°lculo do Retorno Di√°rio do √çndice Local:

$$
\text{√çndice Local}_t = \sum \left( \text{Retorno do Ativo}_i \times \text{Peso Normalizado}_i \right)
$$

---

#### üîπ C√°lculo do √çndice Local Acumulado (como se fosse um pre√ßo):

$$
\text{√çndice Local Acumulado}_t = \prod (1 + \text{√çndice Local}_t)
$$

---

‚úîÔ∏è Esta abordagem permite representar o comportamento agregado do mercado brasileiro ao longo do tempo.

---

Este √≠ndice serve como um term√¥metro do mercado local, podendo ser utilizado tanto para acompanhamento visual quanto como vari√°vel de entrada em modelos de decis√£o quantitativos.
